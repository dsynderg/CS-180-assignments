## Exercise 7: Its Hot and It's Cold
The type column has two values: H='hot' and C='cold'. What is the average rating of each type?

```python
# Write your code to get the average rating by type of cereal below:
hot = cereal[cereal['type']=='H']
cold = cereal[cereal['type']=='C']
display(hot)
display(cold)
```

```python
average_hot = hot['rating'].mean()
average_cold = cold['rating'].mean()
print(f'the hot average is {average_hot} the cold average is {average_cold}')
```
